```python
df_amazon = pd.read_csv("Amazon_merge.csv", encoding="latin1")
df_disney = pd.read_csv("Disney_merge.csv", encoding="latin1")
df_netflix = pd.read_csv("Netflix_merge.csv", encoding="latin1")
IMDB_movies_df = pd.read_csv("IMDB_movie.csv", encoding="latin1")
IMDB_series_df = pd.read_csv("IMDB_serie.csv", encoding="latin1")
```

New dataframes for the each streaming paltform 

- Amazon_IMDB - includes both movies and series, will include columns "Title", "Rating", "Genre", "Year" and "Director"
- Disney_IMDB - includes both movies and series will include columns "Title", "Rating", "Genre", "Year" and "Director"
- Netflix_IMDB - includes both movies and series will include columns "Title", "Rating", "Genre", "Year" and "Director"

Joining dataframes together based on 'Title', if the index matches the row will be added to the fianl dataframe.
Each platform is going to get two data frames, one for movies and one for series. 

left join

- IMDB_series_df, df_amazon
- IMDB_series_df, df_disney
- IMDB_series_df, df_netflix

- IMDB_movies_df, df_amazon
- IMDB_movies_df, df_disney
- IMDB_movies_df, df_netflix

```python
#OLD CODE 
df_IMDB_amazon_series = pd.merge(IMDB_series_df, df_amazon, how='inner', on=["Title"])
df_IMDB_disney_series = pd.merge(IMDB_series_df, df_disney, how='inner', on=["Title"])
df_IMDB_netflix_series = pd.merge(IMDB_series_df, df_netflix, how='inner', on=["Title"])

"""cleaning.inner_merge('test' ,IMDB_series_df, df_amazon, 'inner', 'Title' )
cleaning.inner_merge('test' ,IMDB_series_df, df_disney, 'inner', 'Title' )
cleaning.inner_merge('test' ,IMDB_series_df, df_netflix, 'inner', 'Title' )"""
#OLD CODE
df_IMDB_amazon_movies = pd.merge(IMDB_movies_df, df_amazon, how='inner', on=["Title"])
df_IMDB_disney_movies = pd.merge(IMDB_movies_df, df_disney, how='inner', on=["Title"])
df_IMDB_netflix_movies = pd.merge(IMDB_movies_df, df_netflix, how='inner', on=["Title"])

"""cleaning.merge('test' ,IMDB_movies_df, df_amazon, 'inner', 'Title' )
cleaning.merge('test' ,IMDB_movies_df, df_disney, 'inner', 'Title' )
cleaning.merge('test' ,IMDB_movies_df, df_netflix, 'inner', 'Title' )"""
```

Assign new columns with fixed value to dataframes

Column = "Content type"
Value = "Movie" or "Tv Show"
Column = "Platform"
Value = "Amazon Prime", "Disney Plus", "Netflix"

This allows me to merge the two dataframes into one and reduce the number of dataframes to 1

```python
#Amazon Prime
df_IMDB_amazon_series['Type'] = 'Tv Show'
df_IMDB_amazon_movies['Type'] = 'Movie'
df_IMDB_amazon_series['Platform'] = 'Amazon Prime'
df_IMDB_amazon_movies['Platform'] = 'Amazon Prime'

#Disney Plus
df_IMDB_disney_series['Type'] = 'Tv Show'
df_IMDB_disney_movies['Type'] = 'Movie'
df_IMDB_disney_series['Platform'] = 'Disney Plus'
df_IMDB_disney_movies['Platform'] = 'Disney Plus'

#Netflix
df_IMDB_netflix_series['Type'] = 'Tv Show'
df_IMDB_netflix_movies['Type'] = 'Movie'
df_IMDB_netflix_series['Platform'] = 'Netflix'
df_IMDB_netflix_movies['Platform'] = 'Netflix'
```

Merge series df AND movies df together for each platform.

```python
merge_amazon = [df_IMDB_amazon_movies, df_IMDB_amazon_series]
df_IMDB_amazon = pd.concat(merge_amazon)

merge_disney = [df_IMDB_disney_movies, df_IMDB_disney_series]
df_IMDB_disney = pd.concat(merge_disney)

merge_netflix= [df_IMDB_netflix_movies, df_IMDB_netflix_series]
df_IMDB_netflix = pd.concat(merge_netflix)
```

Export final version dataframes

```python
concat_platforms = [df_IMDB_netflix, df_IMDB_disney, df_IMDB_amazon]
df = pd.concat(concat_platforms)
```

```python
df_IMDB_amazon.to_csv("IMDB_amazon.csv", index=True)
df_IMDB_disney.to_csv("IMDB_disney.csv", index=True)
df_IMDB_netflix.to_csv("IMDB_netflix.csv", index=True)
df.to_csv("df_all.csv", index=True)
```
